It is a crucial step in the data analysis process as it ensures that the data used for analysis is accurate, complete, and consistent.Data cleaning typically involves several steps, including: Handling missing data, Removing duplicates, Standardizing data formats, Handling outliers, Normalizing or scaling data and Data validation

The check_null_values function iterates over each column in a DataFrame to identify and print the count of null values, providing column-specific null counts.

```python
def check_null_values(df_copy):
    """
    Print the count of null values for each column in a DataFrame.

    This function iterates through each column in the DataFrame to check for the presence of null values.
    If a column contains null values, it prints the column name along with the number of null values.

    Parameters:
    df (DataFrame): The pandas DataFrame to check for null values.

    Returns:
    None: This function does not return a value; it only prints information.
    """
    for column in df_copy:
        if df_copy[column].isnull().any():
            print('{0} has {1} null values'.format(column, df_copy[column].isnull().sum()))
```

```python
# Select only numeric columns for calculating average prices
numeric_columns = df_copy.select_dtypes(include=['float64', 'int64'])

# Calculate the average price for each vegetable
average_prices = numeric_columns.mean()

# Identify the most expensive and cheapest vegetables
most_expensive_vegetable = average_prices.idxmax()
cheapest_vegetable = average_prices.idxmin()

# Print the results
print("Most Expensive Vegetable:", most_expensive_vegetable)
print("Cheapest Vegetable:", cheapest_vegetable)
```

Results: Our results shows that the most expensive vegetable over the year of 2023 was Garlic and the cheapest vegetable was Methi.

```python
# Calculate the difference between minimum and maximum prices for each vegetable
bhindi_diff = df_copy['Bhindi (Ladies finger)'].max() - df_copy['Bhindi (Ladies finger)'].min()
tomato_diff = df_copy['Tomato'].max() - df_copy['Tomato'].min()
onion_diff = df_copy['Onion'].max() - df_copy['Onion'].min()
potato_diff = df_copy['Potato'].max() - df_copy['Potato'].min()
brinjal_diff = df_copy['Brinjal'].max() - df_copy['Brinjal'].min()
garlic_diff = df_copy['Garlic'].max() - df_copy['Garlic'].min()
peas_diff = df_copy['Peas'].max() - df_copy['Peas'].min()
methi_diff = df_copy['Methi'].max() - df_copy['Methi'].min()
green_chilli_diff = df_copy['Green Chilli'].max() - df_copy['Green Chilli'].min()
elephant_yam_diff = df_copy['Elephant Yam (Suran)'].max() - df_copy['Elephant Yam (Suran)'].min()

# Printing the results
print("The difference between the minimum and maximum prices per vegetable are as follows:")
print(f"Bhindi = {bhindi_diff};")
print(f"Tomato = {tomato_diff};")
print(f"Onion = {onion_diff};")
print(f"Potato = {potato_diff};")
print(f"Brinjal = {brinjal_diff}, Garlic = {garlic_diff}, Peas = {peas_diff};")
print(f"Methi = {methi_diff};")
print(f"Green Chilli = {green_chilli_diff};")
print(f"Elephant Yam = {elephant_yam_diff}")
```

#### We will show all vegetable prices over the year

```python
import matplotlib.pyplot as plt

# Convert 'Price Dates' column to datetime format if not already done
df_copy['Price Dates'] = pd.to_datetime(df_copy['Price Dates'], format='%d-%m-%Y')

# Sort the DataFrame by 'Price Dates' in ascending order
df_copy = df_copy.sort_values(by='Price Dates')

# List of vegetables
vegetables = ['Bhindi (Ladies finger)', 'Tomato', 'Onion', 'Potato', 'Brinjal',
              'Garlic', 'Peas', 'Methi', 'Green Chilli', 'Elephant Yam (Suran)']

# Loop through each vegetable and plot the line graph
for veg in vegetables:
    plt.figure(figsize=(10, 6))
    plt.plot(df_copy['Price Dates'], df_copy[veg], marker='o', linestyle='-')
    plt.title(f'{veg} Prices Over a Year')
    plt.xlabel('Date')
    plt.ylabel('Price (INR)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

Results: Tomato prices starts off at 18 INR, but remains stable throuhout the year. Onion prices shows a gradual increase throughout the year and peaks in November and shows a decline toward December and significantly drops at the beginning of 2024.

```python
grouped = df_copy.groupby("month")
cols = df_copy.columns[1:-2]  # Exclude 'Price Dates' and 'month' columns
num_cols = min(len(cols), 10)  # Ensure not more than 10 columns are used
index = 0
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(18, 10))
for i in range(2):
    for j in range(5):
        if index < num_cols:
            means = grouped[cols[index]].mean()
            sns.barplot(x=means.index, y=means, ax=axes[i][j])
            for container in axes[i][j].containers:
                axes[i][j].bar_label(container, label_type="center", rotation=90)
            axes[i][j].set_xticklabels(axes[i][j].get_xticklabels(), rotation=90)
        index += 1

plt.tight_layout()
plt.show()
```

```python
index = 0
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(18, 10))
for i in range(2):
    for j in range(5):
        if index < len(cols):
            sns.kdeplot(data=df_copy, x=cols[index], ax=axes[i][j])
            axes[i][j].set_title(cols[index])
        index += 1

plt.tight_layout()
plt.show()
```
